```python
#z = -5/3 + j and plot z on the complex plane.

import numpy as np
import matplotlib.pyplot as plt

# Create a complex number
z = -5/3 + 1j

# label the axes
# a cartesian plane, with boxes
plt.xlabel('Real')
plt.ylabel('Imaginary')

# plot the complex number
plt.plot(z.real, z.imag, 'o')

# save the plot
plt.savefig('q1a.png')
```
